# STEP 4 — Similarity Engine

```python
import pandas as pd
import pickle
from sklearn.metrics.pairwise import cosine_similarity


# ---------------------------
# LOAD DATA + MODELS
# ---------------------------
df = pd.read_csv("src data/shl_feature_engineered.csv")

with open("tfidf pkl/tfidf_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

with open("tfidf pkl/tfidf_matrix.pkl", "rb") as f:
    tfidf_matrix = pickle.load(f)

```

```python

# ---------------------------
# RECOMMEND FUNCTION
# ---------------------------
def recommend_assessments(user_input, top_n=5):

    # convert user input into vector
    user_vector = vectorizer.transform([user_input])

    # calculate similarity
    similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

    # get top matches
    scores = similarity_scores.flatten()

    top_indices = scores.argsort()[::-1][:top_n]

    recommendations = df.iloc[top_indices][
        ["Assessment", "Role", "Skills", "Duration"]
    ]

    return recommendations

```

```python

# ---------------------------
# TEST ENGINE
# ---------------------------
if __name__ == "__main__":

    user_profile = "data scientist python ml nlp senior technical it"

    results = recommend_assessments(user_profile)

    print("\nRecommended Assessments:\n")
    print(results)
```
