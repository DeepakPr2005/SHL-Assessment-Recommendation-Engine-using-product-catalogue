```python
import pandas as pd

# ---------------------------
# LOAD DATA
# ---------------------------
df = pd.read_csv("src data/shl_catalogue.csv")

print("Original Shape:", df.shape)

```

```python
# ---------------------------
# STANDARDIZE TEXT
text_columns = [
    "Assessment",
    "Role",
    "Skills",
    "Category",
    "Level",
    "Test_Type",
    "Industry"
]

for col in text_columns:
    df[col] = df[col].astype(str).str.lower().str.strip()

```

```python

# ---------------------------
# CLEAN SKILLS COLUMN
# Convert space → comma

df["Skills"] = df["Skills"].apply(lambda x: ",".join(x.split()))

```

```python

# ---------------------------
# CONVERT YES / NO → 1 / 0
# ---------------------------
df["Remote_Testing"] = df["Remote_Testing"].map({"Yes": 1, "No": 0})
df["Adaptive"] = df["Adaptive"].map({"Yes": 1, "No": 0})

```

```python

# ---------------------------
# ENCODE DIFFICULTY
# ---------------------------
difficulty_map = {
    "easy": 1,
    "medium": 2,
    "hard": 3
}

df["Difficulty"] = df["Difficulty"].str.lower().map(difficulty_map)

```

```python

# ---------------------------
# REMOVE DUPLICATES
# ---------------------------
df = df.drop_duplicates()

```

```python

# ---------------------------
# HANDLE MISSING VALUES
# ---------------------------
df = df.dropna()

```

```python

# ---------------------------
# SAVE CLEAN DATA
# ---------------------------
df.to_csv("src data/shl_catalogue_cleaned.csv", index=False)

print("Data Cleaning Completed ✅")
print("Cleaned Shape:", df.shape)
```
