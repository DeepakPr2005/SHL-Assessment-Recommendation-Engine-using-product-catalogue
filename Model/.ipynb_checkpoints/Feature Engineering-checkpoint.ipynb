# STEP 2 — Feature Engineering (Second Coding Step)

```python
# ---------------------------
# LOAD CLEAN DATA
# ---------------------------

import pandas as pd

df = pd.read_csv("shl_catalogue_cleaned.csv")

print("Dataset Loaded:", df.shape)

```

```python
# ---------------------------
# CREATE COMBINED FEATURES
# ---------------------------


df["combined_features"] = (
    df["Role"] + " " +
    df["Skills"] + " " +
    df["Category"] + " " +
    df["Level"] + " " +
    df["Industry"]
)

```

```python
# ---------------------------
# OPTIONAL NUMERIC FEATURES
# ---------------------------


df["meta_features"] = (
    df["Difficulty"].astype(str) + " " +
    df["Duration"].astype(str)
)

```

```python
# ---------------------------
# FINAL FEATURE COLUMN
# ---------------------------


df["final_features"] = (
    df["combined_features"] + " " +
    df["meta_features"]
)

```

```python
# ---------------------------
# SAVE FEATURE DATASET
# ---------------------------



df.to_csv("shl_feature_engineered.csv", index=False)

print("Feature Engineering Completed")
```
